```python
import numpy as np
import matplotlib.pyplot as plt 

from skimage.color import rgb2gray
from sklearn.cluster import KMeans
from skimage.io import imread, imsave
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
imgPath = 'ultrassound/cropped/1100212039.bmp'

imgTest = imread(imgPath)
plt.imshow(imgTest, cmap='gray')
plt.axis('off')
plt.show()
```

```python
def applyKmeans(img, k):
    img = img.copy()

    km = KMeans(n_clusters=k)
    
    if len(img.shape) == 3:
        img = rgb2gray(img)
    
    imgShape = img.shape
    imgReshape = np.reshape(img, (-1, 1))
    result = km.fit(imgReshape)

    imgReshape = np.reshape(result.labels_, imgShape)
    return result, imgReshape
```

```python
def maxAndMinLabels(cluster):
    result = list(enumerate(cluster))
    def reverse(a: tuple):
        return (a[1], a[0])
    result = list(map(reverse, result))
    result.sort()
    return result[0][1], result[-1][1]
```

```python
result, imgReshape = applyKmeans(imgTest, 4)
```

```python
plt.imshow(imgReshape)
plt.axis('off')
plt.show()
```

```python
minValue, maxValue, = maxAndMinLabels(result.cluster_centers_)
mask = (imgReshape != minValue) & (imgReshape != maxValue)

plt.imshow(mask)
plt.axis('off')
plt.show()
```

```python
imgResult = np.zeros(imgTest.shape)

imgResult[mask] = imgTest[mask]

plt.imshow(imgResult)
plt.axis('off')
plt.show()
```

```python
result, imgReshape = applyKmeans(imgResult, 5)
```

```python
minValue, maxValue, = maxAndMinLabels(result.cluster_centers_)
mask = (imgReshape != minValue) & (imgReshape != maxValue)

plt.imshow(mask)
plt.axis('off')
plt.show()
```
